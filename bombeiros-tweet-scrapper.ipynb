```python
import GetOldTweets3 as got
from datetime import datetime
import csv

#Parametros
username = 'bvsc_joinville'
count = 3000

# Cria objeto de busca
tweetCriteria = got.manager.TweetCriteria().setUsername(username)\
                                        .setMaxTweets(count)
# Cria lista com tweets
tweets = got.manager.TweetManager.getTweets(tweetCriteria)

# Cria lista com os campos selecionados
user_tweets = [[tweet.date, tweet.text, tweet.mentions, tweet.hashtags, tweet.favorites, tweet.retweets, tweet.to, tweet.id, tweet.permalink] for tweet in tweets]

# Converte data
for i in user_tweets:
    i[0] = i[0].strftime("%d-%m-%Y - %H:%M:%S")

# Acompanha progresso
print("o número de itens na lista é: " + str(len(user_tweets)))
```

```python
#Nome do arquivo de saída

file_name = 'twitter-bombeiros-db-atualizado-julho-2020.csv'

#Abrir CSV, criar writer e cabeçalho

file = open(file_name, 'w', encoding='utf8', newline='')
writer = csv.writer(file, delimiter='|')
writer.writerow(['data', 'texto', 'mentions', 'hashtags', 'favoritos', 'retweets', 'to', 'tweetId', 'link'])

#Escrever na planilha
for item in user_tweets:
    data = item[0]
    texto = item[1] 
    mentions = item[2]
    hashtags = item[3]
    favoritos = item[4]
    retweets = item[5] 
    to = item[6]
    tweetId = item[7]
    link = item[8]
    
    rowData = [data, texto, mentions, hashtags, favoritos, retweets, to, tweetId, link]
    print(rowData) 
    writer.writerow(rowData)

file.close()
```
